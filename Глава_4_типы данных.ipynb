```python
import pandas as pd

d = {'sales': [100, 150, 150, 120, 135, 180, 230, 500, 501, 499, 500, 500]}
df = pd.DataFrame(d)


def evalInc(data):
    arr = pd.Series([np.nan])
    data['inc'] =  pd.concat([arr , pd.Series(np.diff(data.sales))]).reset_index(drop=True).fillna(0).astype(int)
    return data
```
